```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.genfromtxt("Afile.txt", delimiter=",")
```

```python
N=101
X=np.ndarray((N,N,N),dtype=float)
Y=np.ndarray((N,N,N),dtype=float)
Z=np.ndarray((N,N,N),dtype=float)
Ax=np.ndarray((N,N,N),dtype=float)
Ay=np.ndarray((N,N,N),dtype=float)
Az=np.ndarray((N,N,N),dtype=float)
Bx=np.ndarray((N-1,N-1,N-1),dtype=float)
By=np.ndarray((N-1,N-1,N-1),dtype=float)
```

```python
for i in range(N):
    for j in range(N):
        for k in range(N):
            X[i,j,k] = A[N*N*i+N*j+k,0]
            Y[i,j,k] = A[N*N*i+N*j+k,1]
            Z[i,j,k] = A[N*N*i+N*j+k,2]
            Ax[i,j,k] = A[N*N*i+N*j+k,3]
            Ay[i,j,k] = A[N*N*i+N*j+k,4]
            Az[i,j,k] = A[N*N*i+N*j+k,5]
```

```python
for i in range(N-1):
    for j in range(N-1):
        for k in range(N-1):
            Bx[i,j,k] = (Ay[i,j,k+1] - Ay[i,j,k])/(Z[i,j,k+1] - Z[i,j,k]) - (Az[i,j+1,k] - Az[i,j,k])/(Y[i,j+1,k] - Y[i,j,k])
            By[i,j,k] = -(Ax[i,j,k+1] - Ax[i,j,k])/(Z[i,j,k+1] - Z[i,j,k]) + (Az[i+1,j,k] - Az[i,j,k])/(X[i+1,j,k] - X[i,j,k])
```

```python
fig, axs = plt.subplots(1,2, figsize=(10, 5))
axs[0].streamplot(Y[:-1,:-1,50],X[:-1,:-1,50],By[:,:,50], Bx[:,:,50])
axs[0].set_title("B-field streamlines")
CS = axs[1].contourf(Y[:-1,:-1,50],X[:-1,:-1,50],np.sqrt(By[:,:,50]**2+Bx[:,:,50]**2),levels = [0,1,2,3,4,5,6,7,8])
axs[1].set_title("B-field magnitude")
cbar = fig.colorbar(CS)
plt.savefig("with_mu1000.png", dpi=150)
```
